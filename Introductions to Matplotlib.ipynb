# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2.Setup plot
fig, ax = plt.subplots(figsize = (5,10))
    
# 3.Plot data
ax.plot(x,y)
# 4. Customize plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")
# 5. Save & show (you save the figure)
fig.savefig("matplotlib-anatomy-of-a-plot.png")
```

```python
# Two options for subplots
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10,5))
# Plot to each different axis

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Two options for subplots
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                                    ncols=2,
                                                    figsize=(10,5))
# Plot to each different axis

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.hist(np.radians(1000))
ax4.bar(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python

# Subplots option 2
fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each different axis
ax[0, 0].plot(x,x/2);
ax[0, 1].scatter(np.random.random(10),np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1, 1].hist(np.radians(1000));
```

## Wich one should you see? (pyplot vs matplotlib 00 method?)

##### When plotting something quickly, okey to use the pyplot method
##### when plotting somthing more advanced, use the OO method


```python
heart_disase.head()
```

```python
over_50 = heart_disase[heart_disase["age"]>50]
over_50.head()
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',);
```

```python
# 00 method
fig,ax = plt.subplots(figsize =(20,6))
over_50.plot(kind ='scatter',
             x='age',
             y='chol',
             c='target',
            ax=ax);

ax.set_xlim([45,100]);
```

```python
over_50.target
```

```python
# oo ethod from scratch
fig, ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
# Customize the plot
ax.set(title="Heart Disase and Cholestrol Levels")
# add a legend
ax.legend(*scatter.legend_elements(), title="Target");
# Add a horizonal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach
fig,(ax0,ax1) = plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disase and cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")
# add a legand to ax0
ax0.legend(*scatter.legend_elements(),title="Target")
# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])

#Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

#add a legand to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle='--');
## add a little to the figure
fig.suptitle("Heart Diease Analysis", fontsize=16, fontweight="bold");
```

# Changing the cmap

```python
plt.style.use('seaborn-whitegrid')
```

```python
# No cmap change
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"],
                    cmap="plasma")
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.axhline(y=over_50["chol"].mean(), 
           c='b', 
           linestyle='--', 
           label="Average");
ax.legend(*scatter.legend_elements(), title="Target");
```

```python
#Customizing the y and x axis limitations
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 10))
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Set the x axis
ax0.set_xlim([50, 80])

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax0.legend(*scatter.legend_elements(), title="Target")

# Axis 1, 1 (row 1, column 1)
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Set the y axis
ax1.set_ylim([60, 200])

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
