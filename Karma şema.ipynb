```python
import matplotlib.pyplot as plt
yaslar = [12,2,23,23,44,12,43,18,17,43,88,90]

yas_gruplar = [0,10,20,30,40,50,60,70]

plt.hist(yaslar,yas_gruplar,histtype ="bar", rwidth=0.6)
plt.xlabel("Yaş Groupları")
plt.ylabel("Kişi Sayısı")
plt.title("Histogram Grafiği")

plt.show()
```

```python
plt.bar([0.25,0.50,0.75,1.00],[2,3,5,2],label ="BMW", width=4)
plt.bar([1.25,1.50,1.75,2.00],[5,2,9,1],label ="Audi",width =3)

plt.legend()
plt.xlabel("Gün")
plt.ylabel("Mesafe(km)")

plt.title("Araç Bilgileri")

plt.show()
```

```python
goal_types = "Penaltı","Kaleye Atılan Şut","Serbest Vuruş"

goals = [12,20,18]

colors = ["y","r","b"]

plt.pie(goals,labels = goal_types, colors=colors,shadow=True,explode =(0.05,0.05,0.05))

plt.show()
```
