```python
import matplotlib.pyplot as plt

yil = [2011,2012,2013,2014,2015] # Yıl ile Oyuncu sayısı aynı miktarda olsun sakın unutma hata alırsın yoksa...

oyuncu1 = [8,10,12,7,9]
oyuncu2 = [8,4,21,3,5]
oyuncu3 = [4,23,4,5,2]

#Stack Plot

plt.plot([],[],color ="y",label = "Oyuncu1")
plt.plot([],[],color ="r",label = "Oyuncu2")
plt.plot([],[],color ="b",label = "Oyuncu3")
plt.title("Yıllara Göre Atılan Gol Sayısı")

plt.xlabel("Yıllar")
plt.ylabel("Gol Sayısı")

plt.stackplot(yil,oyuncu1,oyuncu2,oyuncu3, colors=["y","r","b"] )
plt.legend()

plt.show()
```

```python
goal_types = "Penaltı","Kaleye Atılan Şut","Serbest Vuruş"

goals = [12,20,5]

colors =['y','r','b']

plt.pie(goals,labels = goal_types,colors=colors, shadow =True, explode =(0.05,0.05,0.05))
plt.show()
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20], label="BMW",width=10)
plt.bar([0.75,1.75,2.75,3.75,75],[80,20,20,50,60], plt.legendA(="Audi",width =10)

plt.legend()
plt.xlabel("Gün")
plt.ylabel("Mesafe(km)")
plt.title ("Araç Bilgileri")

plt.legend()
plt.show()
```
