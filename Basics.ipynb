```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
y = [5,6,8,9]

plt.plot(x,y)

plt.axis([0,6,0,20])

plt.show()
```

```python
plt.title("Grafik BaÅlÄ±k") #Ana BaÅlÄ±k
plt.xlabel("Fiyatlar") # Y Ã§ubuÄu BaÅlÄ±ÄÄ±
plt.ylabel("SatÄ±lcaklar")# X Ã§ubuÄu BaÅlÄ±ÄÄ±
```

```python
plt.plot(x,y,color="red",linewidth="5") #EÄer baÅka bir renk kullanmak isyorsan bu kod dizini kullanÄ±lacak.
```

```python
plt.plot(x,y,"o--g") #DÃ¼z bir yeÅi Ã§izgi- eÄer fazladan bir tane daha eksi koyarsan iki Ã§izgi olur.
plt.show()
```

```python
x = np.linspace(0,2,100) #sÄ±fÄ±rdan baÅla 2 ye kadar git 100 tane ara sayÄ± gÃ¶ster
plt.plot(x, x, label="linear")
plt.plot(x, x**2, label = "quadratic")
plt.plot(x, x**3, label="cubic")

plt.xlabel("x label")
plt.ylabel("y label")

plt.legend()

plt.title("Label Tabl")

plt.show()
```

```python
x = np.linspace(0,2,100)
fig,axs = plt.subplots(3) #3  tane axes oluÅturmamÄ±zÄ± saÄlÄ±yor. yani 3 ayrÄ± tablo

axs[0].plot(x , x, color ="red")
axs[0].set_title("linear")

axs[1].plot(x , x**2, color ="green")
axs[1].set_title("quadratic")

axs[2].plot(x , x**3, color ="yellow")
axs[2].set_title("cubic")

plt.tight_layout() #Her Åe iÃ§ iÃ§e geÃ§ti baÅlÄ±klarda dÃ¼zeltmek iÃ§in 

plt.show()
```

```python
x = np.linspace(0,2,100)
fig,axs = plt.subplots(2,2)

fig.suptitle("GRafik BaÅlÄ±ÄÄ±")
axs[0,0].plot(x , x, color ="red")
axs[0,1].plot(x , x**2, color ="yellow")
axs[1,0].plot(x , x**3, color ="green")
axs[1,1].plot(x , x**4, color ="blue")


plt.show()
```
