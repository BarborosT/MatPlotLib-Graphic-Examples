```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
y = [5,6,8,9]

plt.plot(x,y)

plt.axis([0,6,0,20])

plt.show()
```

```python
plt.title("Grafik Başlık") #Ana Başlık
plt.xlabel("Fiyatlar") # Y çubuğu Başlığı
plt.ylabel("Satılcaklar")# X çubuğu Başlığı
```

```python
plt.plot(x,y,color="red",linewidth="5") #Eğer başka bir renk kullanmak isyorsan bu kod dizini kullanılacak.
```

```python
plt.plot(x,y,"o--g") #Düz bir yeşi çizgi- eğer fazladan bir tane daha eksi koyarsan iki çizgi olur.
plt.show()
```

```python
plt.plot(x,y,"o--r")
plt.show()
```

```python
x = np.linspace(0,2,100) #sıfırdan başla 2 ye kadar git 100 tane ara sayı göster
plt.plot(x, x, label="linear")
plt.plot(x, x**2, label = "quadratic")
plt.plot(x, x**3, label="cubic")

plt.xlabel("x label")
plt.ylabel("y label")

plt.legend()

plt.title("Label Tabl")

plt.show()
```

```python
x = np.linspace(0,2,100)
fig,axs = plt.subplots(3) #3  tane axes oluşturmamızı sağlıyor. yani 3 ayrı tablo

axs[0].plot(x , x, color ="red")
axs[0].set_title("linear")

axs[1].plot(x , x**2, color ="green")
axs[1].set_title("quadratic")

axs[2].plot(x , x**3, color ="yellow")
axs[2].set_title("cubic")

plt.tight_layout() #Her şe iç içe geçti başlıklarda düzeltmek için 

plt.show()
```

```python
x = np.linspace(0,2,100)
fig,axs = plt.subplots(2,2)

fig.suptitle("GRafik Başlığı")
axs[0,0].plot(x , x, color ="red")
axs[0,1].plot(x , x**2, color ="yellow")
axs[1,0].plot(x , x**3, color ="green")
axs[1,1].plot(x , x**4, color ="blue")


plt.show()
```
