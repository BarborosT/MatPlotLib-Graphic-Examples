```python
import matplotlib.pyplot as plt

yıl=[2011,2012,2013,2014,2015]

oyuncu1=[8,10,12,7,9]
oyuncu2=[8,4,21,3,2]
oyuncu3=[4,23,4,5,2]

plt.plot([],[], color ="y",label ="Oyuncu1")
plt.plot([],[], color ="r",label ="Oyuncu2")
plt.plot([],[], color ="b", label ="Oyuncu3")
plt.title("Yıllara göre atılan gol sayısı")

plt.xlabel("Yıllar")
plt.ylabel("Gol Sayısı")

plt.stackplot(yıl,oyuncu1,oyuncu2,oyuncu3, color=["y","r","b"])
plt.legend()

plt.show()
```
